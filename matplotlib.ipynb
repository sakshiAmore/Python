# Data visualization
Data visualisation is representation of data in graphical form.

Matplotlib contains all the functions which are responsible to form a graph in python.


```python
#importing libraries for visualisation
import matplotlib as plt
from matplotlib import pyplot as plt #command to tell python that you want to use matplotlib
from matplotlib import style
```

```python
plt.plot([1,2,3],[4,5,1])
plt.show()
```

```python
plt.plot([2,4,6,7,9],[5,8,3,6,9])
plt.show()
```

```python
style.use('ggplot')
x=[1,2,3]#(1,1)(2,2)(3,3)
y=[1,2,3]
x1=[11,2,3]  #(11,1)(2,2)(3,3)
y1=[1,2,3]
plt.plot(x,y,'g',label='purchased')
plt.plot(x1,y1,'c',label='sold')
plt.title('Milk info')
plt.ylabel('sold')
plt.xlabel('purchased')
plt.legend()
plt.show()
```

```python
x=[2,3,6]
y=[3,7,9]
x1=[2,8,6]
y1=[4,7,9]
x2=[3,5,8]
y2=[6,9,7]
plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.title('Toys')
plt.xlabel('Sold toys')
plt.ylabel('Purchased Toys')

plt.show()
```

```python
#GGPLOT
style.use('ggplot')#name of the plot is gglot
x=[5,8,10]#coordinates for x axes in first plot 
y=[12,16,6]#coordinates for y axes in first plot
x2=[6,9,11]#coordinates for x axes in second plot
y2=[6,15,7]#coordinates for y axes in second plot
```

```python
plt.plot(x,y,'g',label='Milk',linewidth=5)#label is name of plot and linewidth is the thickness of the plot
plt.plot(x2,y2,'c',label='Bread',linewidth=5)
plt.title('Milk vs Bread') #title of the graph
plt.ylabel('Milk')#title for y axies
plt.xlabel('Bread')#title for x axies
plt.legend()#indicator shown on the top right of the graph
plt.grid(False,color='b')#background grid lines
plt.show()#mandatory command to display any graph
```

Q. show a graph on distance and time

```python
style.use('ggplot')
x=[2,4,6,7,9]
y=[1,3,5,6,8]
plt.plot(x,'r',label='Distance',linewidth=2)
plt.plot(y,'y',label='Time',linewidth=2)
plt.title('Distance and Time')
plt.xlabel('Distance')
plt.ylabel('Time')
plt.legend()
plt.grid(True, color='g')
plt.show()
```

```python
#Bar graph
import matplotlib.pyplot as plt #importing libraries
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="population of youth")#giving x and y coordinates.
plt.bar([2,4,6,8,10],[8,6,2,5,6],label="population of senior citizen")
plt.xlabel('population of youth')
plt.ylabel('population of senior citizen')
plt.title('population')
plt.legend()
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt#importing libraries
population_ages=[22,55,62,34,56,78,89,98,12,34,55,77,88,99,22,36,56,101,50,46,21,43,48]#categorical values in the graph
bins=[0,10,20,40,50,60,70,80,90,100,110]#creating bins
plt.hist(population_ages,histtype='bar',rwidth=0.8)#in histogram we use rwidth
plt.xlabel=('x')
plt.ylabel=('y')
plt.title('histogram')
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt#importing libraries
population_ages=[22,55,62,34,56,78,89,98,12,34,55,77,88,99,22,36,56,101,50,46,21,43,48]#categorical values in the graph
bins=[0,10,20,40,50,60,70,80,90,100,110]#creating bins
plt.hist(population_ages,bins,histtype='bar',rwidth=0.8)
#plt.hist(bins,histtype='bar',rwidth=0.8)  #in histogram we use rwidth
plt.xlabel=('x')
plt.ylabel=('y')
plt.title('histogram')
plt.show()
```

```python
#Scatter plot
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]
plt.scatter(x,y,label='skits',color='r',s=30,marker='2')
plt.xlabel('Advertising though social media')
plt.ylabel('Advertising though newspaper')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

```python
#Piechart
import matplotlib.pyplot as plt#importing library
slices=[7,2,2,13]#plotting values in pie chart
activities=['sleeping','eating','working','playing'] #labels
cols=['y','m','r','b']#colours
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct='%0.2f%%')#pctdistance=1)
plt.title('Pie Plot')
plt.show()
#slices is a list variable consisting of all the datat to be plotted in the pie chart
#activities is a list variable consisting of all the labels to be repres
```
